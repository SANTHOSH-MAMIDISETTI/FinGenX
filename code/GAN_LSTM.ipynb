```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from torch.utils.data import DataLoader, TensorDataset

# Load the dataset
df = pd.read_csv('/content/ADANIPORTS.csv', parse_dates=['Date'])

# Ensure the 'Close' column exists
if 'Close' not in df.columns:
    raise ValueError("The dataset does not contain a 'Close' column.")

# Sort by date in case the data is not in chronological order
df = df.sort_values('Date')

# Select the 'Close' price column
close_prices = df['Close'].values.reshape(-1, 1)

# Normalize the 'Close' prices
scaler = MinMaxScaler(feature_range=(0, 1))
normalized_prices = scaler.fit_transform(close_prices)

# Function to create sequences
def create_sequences(data, sequence_length):
    sequences = []
    targets = []
    for i in range(len(data) - sequence_length):
        seq = data[i:i + sequence_length]
        target = data[i + sequence_length]
        sequences.append(seq)
        targets.append(target)
    sequences = np.array(sequences).reshape(-1, sequence_length, 1)
    targets = np.array(targets)
    return (torch.tensor(sequences, dtype=torch.float32),
            torch.tensor(targets, dtype=torch.float32))

# Hyperparameters
sequence_length = 10
batch_size = 32
noise_dim = 100

# Create sequences
sequences, targets = create_sequences(normalized_prices, sequence_length)
dataset = TensorDataset(sequences, targets)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

class LSTMGenerator(nn.Module):
    def __init__(self, noise_dim, hidden_dim=256, output_dim=1, num_layers=3, seq_len=10, num_heads=2):
        super(LSTMGenerator, self).__init__()
        self.noise_dim = noise_dim
        self.seq_len = seq_len
        self.hidden_dim = hidden_dim

        assert hidden_dim % num_heads == 0, "hidden_dim must be divisible by num_heads"

        self.lstm = nn.LSTM(noise_dim, hidden_dim, num_layers, batch_first=True, dropout=0.3)
        self.attention = nn.MultiheadAttention(embed_dim=hidden_dim, num_heads=num_heads, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
        self.residual_fc = nn.Linear(noise_dim, output_dim)
        self.tanh = nn.Tanh()

    def forward(self, noise):
        batch_size = noise.size(0)
        noise_seq = noise.unsqueeze(1).expand(-1, self.seq_len, -1)

        lstm_out, _ = self.lstm(noise_seq)
        attn_out, _ = self.attention(lstm_out, lstm_out, lstm_out)

        combined_out = attn_out + lstm_out
        seq_output = self.fc(combined_out)

        residual_output = self.residual_fc(noise).unsqueeze(1).expand(-1, self.seq_len, -1)
        final_output = self.tanh(seq_output + residual_output)

        return final_output


class LSTMDiscriminator(nn.Module):
    def __init__(self, input_dim, hidden_dim=256, num_layers=3):
        super(LSTMDiscriminator, self).__init__()
        self.bilstm = nn.LSTM(input_dim, hidden_dim, num_layers,
                             bidirectional=True, batch_first=True, dropout=0.3)
        self.fc = nn.utils.spectral_norm(nn.Linear(hidden_dim * 2, 1))
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        bilstm_out, _ = self.bilstm(x)
        last_hidden_state = bilstm_out[:, -1, :]
        output = self.fc(last_hidden_state)
        return self.sigmoid(output)


# Train GAN
# Train GAN
def train_gan(generator, discriminator, dataloader, num_epochs, g_optimizer, d_optimizer, criterion, device, noise_dim):
    d_losses = []
    g_losses = []

    for epoch in range(num_epochs):
        total_d_loss = 0
        total_g_loss = 0
        num_batches = 0

        for real_sequences, _ in dataloader:
            batch_size = real_sequences.size(0)
            real_sequences = real_sequences.to(device)

            # Train Discriminator
            d_optimizer.zero_grad()
            noise = torch.randn(batch_size, noise_dim).to(device)

            # Generate fake sequences and ensure that the generator output requires gradients
            fake_sequences = generator(noise)

            # Ensure fake_sequences requires gradients
            fake_sequences = fake_sequences.detach().requires_grad_()

            real_outputs = discriminator(real_sequences)
            fake_outputs = discriminator(fake_sequences)

            d_loss_real = criterion(real_outputs, torch.ones_like(real_outputs))
            d_loss_fake = criterion(fake_outputs, torch.zeros_like(fake_outputs))
            d_loss = (d_loss_real + d_loss_fake) / 2

            d_loss.backward()
            d_optimizer.step()

            # Train Generator
            g_optimizer.zero_grad()
            fake_outputs = discriminator(fake_sequences)
            g_loss = criterion(fake_outputs, torch.ones_like(fake_outputs))

            g_loss.backward()
            g_optimizer.step()

            total_d_loss += d_loss.item()
            total_g_loss += g_loss.item()
            num_batches += 1

        # Calculate and store average losses
        avg_d_loss = total_d_loss / num_batches
        avg_g_loss = total_g_loss / num_batches
        d_losses.append(avg_d_loss)
        g_losses.append(avg_g_loss)

        # Print progress
        print(f"Epoch [{epoch+1}/{num_epochs}] - D Loss: {avg_d_loss:.4f}, G Loss: {avg_g_loss:.4f}")

        # Plot progress every 100 epochs
        if (epoch + 1) % 100 == 0:
            plt.figure(figsize=(10, 5))
            plt.plot(d_losses, label='Discriminator Loss')
            plt.plot(g_losses, label='Generator Loss')
            plt.xlabel('Epoch')
            plt.ylabel('Loss')
            plt.title(f'GAN Training Progress - Epoch {epoch+1}')
            plt.legend()
            plt.grid(True)
            plt.show()

    return d_losses, g_losses



# Setup and training
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
hidden_dim = 56
num_heads = 2

# Initialize models
generator = LSTMGenerator(noise_dim=noise_dim, hidden_dim=hidden_dim, num_heads=num_heads).to(device)
discriminator = LSTMDiscriminator(input_dim=1, hidden_dim=hidden_dim).to(device)

# Setup optimizers
learning_rate = 0.01
g_optimizer = optim.Adam(generator.parameters(), lr=learning_rate)
d_optimizer = optim.Adam(discriminator.parameters(), lr=learning_rate)
criterion = nn.BCELoss()

# Train the GAN
num_epochs = 500
d_losses, g_losses = train_gan(generator, discriminator, dataloader, num_epochs,
                              g_optimizer, d_optimizer, criterion, device, noise_dim)

# Plot final losses
plt.figure(figsize=(12, 6))
plt.plot(d_losses, label='Discriminator Loss', alpha=0.7)
plt.plot(g_losses, label='Generator Loss', alpha=0.7)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('GAN Training Loss Over Time')
plt.legend()
plt.grid(True)
plt.show()

# Save losses to CSV
loss_df = pd.DataFrame({
    'Epoch': range(1, num_epochs + 1),
    'Discriminator_Loss': d_losses,
    'Generator_Loss': g_losses
})
loss_df.to_csv('gan_training_losses.csv', index=False)
print("\nLosses have been saved to 'gan_training_losses.csv'")

# Generate some sample sequences
with torch.no_grad():
    test_noise = torch.randn(5, noise_dim).to(device)
    generated_sequences = generator(test_noise).cpu().numpy()

# Plot some sample generated sequences
plt.figure(figsize=(12, 6))
for i in range(5):
    plt.plot(generated_sequences[i, :, 0], label=f'Sequence {i+1}')
plt.xlabel('Time Step')
plt.ylabel('Generated Value')
plt.title('Sample Generated Sequences')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from torch.utils.data import DataLoader, TensorDataset

# Load the dataset
df = pd.read_csv('/content/ADANIPORTS.csv', parse_dates=['Date'])

# Ensure the 'Close' column exists
if 'Close' not in df.columns:
    raise ValueError("The dataset does not contain a 'Close' column.")

# Sort by date in case the data is not in chronological order
df = df.sort_values('Date')

# Select the 'Close' price column
close_prices = df['Close'].values.reshape(-1, 1)

# Normalize the 'Close' prices
scaler = MinMaxScaler(feature_range=(0, 1))
normalized_prices = scaler.fit_transform(close_prices)

# Function to create sequences
def create_sequences(data, sequence_length):
    sequences = []
    targets = []
    for i in range(len(data) - sequence_length):
        seq = data[i:i + sequence_length]
        target = data[i + sequence_length]
        sequences.append(seq)
        targets.append(target)
    sequences = np.array(sequences).reshape(-1, sequence_length, 1)
    targets = np.array(targets)
    return (torch.tensor(sequences, dtype=torch.float32),
            torch.tensor(targets, dtype=torch.float32))

# Hyperparameters
sequence_length = 10
batch_size = 32
noise_dim = 100

# Create sequences
sequences, targets = create_sequences(normalized_prices, sequence_length)
dataset = TensorDataset(sequences, targets)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

import torch
import torch.nn as nn

class LSTMGenerator(nn.Module):
    def __init__(self, noise_dim, hidden_dim=256, output_dim=1, num_layers=3, seq_len=10, num_heads=2, num_features=1):
        super(LSTMGenerator, self).__init__()
        self.noise_dim = noise_dim
        self.seq_len = seq_len
        self.hidden_dim = hidden_dim
        self.num_features = num_features  # Number of features in the output sequence

        assert hidden_dim % num_heads == 0, "hidden_dim must be divisible by num_heads"

        # Define layers
        self.lstm = nn.LSTM(noise_dim, hidden_dim, num_layers, batch_first=True, dropout=0.3)
        self.attention = nn.MultiheadAttention(embed_dim=hidden_dim, num_heads=num_heads, batch_first=True)
        self.fc = nn.Linear(hidden_dim, num_features)  # Output dimension should match num_features (15)
        self.residual_fc = nn.Linear(noise_dim, num_features)  # Residual connection output should match num_features
        self.batch_norm = nn.BatchNorm1d(hidden_dim)
        self.tanh = nn.Tanh()

    def forward(self, noise):
        batch_size = noise.size(0)

        # Expand noise to create a sequence for each batch
        # We now generate sequences with 15 features per time step, so we need to adjust the noise shape
        noise_seq = noise.unsqueeze(1).expand(-1, self.seq_len, self.noise_dim)

        # Pass through LSTM
        lstm_out, _ = self.lstm(noise_seq)

        # Apply multihead attention
        attn_out, _ = self.attention(lstm_out, lstm_out, lstm_out)

        # Apply batch normalization
        attn_out = self.batch_norm(attn_out.reshape(-1, attn_out.size(-1))).reshape(attn_out.size())

        # Combine LSTM output and attention output (residual connection)
        combined_out = attn_out + lstm_out

        # Generate sequence output with the desired number of features (15)
        seq_output = self.fc(combined_out)

        # Apply residual connection and tanh activation
        residual_output = self.residual_fc(noise).unsqueeze(1).expand(-1, self.seq_len, -1)
        final_output = self.tanh(seq_output + residual_output)

        return final_output



class LSTMDiscriminator(nn.Module):
    def __init__(self, input_dim, hidden_dim=256, num_layers=3):
        super(LSTMDiscriminator, self).__init__()
        self.bilstm = nn.LSTM(input_dim, hidden_dim, num_layers,
                             bidirectional=True, batch_first=True, dropout=0.3)
        self.fc = nn.utils.spectral_norm(nn.Linear(hidden_dim * 2, 1))  # Output dimension of 1 (for classification)
        self.batch_norm = nn.BatchNorm1d(hidden_dim * 2)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        bilstm_out, _ = self.bilstm(x)
        last_hidden_state = bilstm_out[:, -1, :]

        # Apply batch normalization
        norm_out = self.batch_norm(last_hidden_state)
        output = self.fc(norm_out)
        return self.sigmoid(output)


# Setup and training
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
hidden_dim = 56
num_heads = 2

# Initialize models
generator = LSTMGenerator(noise_dim=noise_dim, hidden_dim=hidden_dim, num_heads=num_heads).to(device)
discriminator = LSTMDiscriminator(input_dim=1, hidden_dim=hidden_dim).to(device)

# Setup optimizers with smaller learning rate
learning_rate = 0.0002
g_optimizer = optim.Adam(generator.parameters(), lr=learning_rate, betas=(0.5, 0.999))
d_optimizer = optim.Adam(discriminator.parameters(), lr=learning_rate, betas=(0.5, 0.999))
criterion = nn.BCEWithLogitsLoss()  # More numerically stable loss

# Modify Discriminator's loss calculation with label smoothing
def train_gan(generator, discriminator, dataloader, num_epochs, g_optimizer, d_optimizer, criterion, device, noise_dim):
    d_losses = []
    g_losses = []

    for epoch in range(num_epochs):
        total_d_loss = 0
        total_g_loss = 0
        num_batches = 0

        for real_sequences, _ in dataloader:
            batch_size = real_sequences.size(0)
            real_sequences = real_sequences.to(device)

            # Train Discriminator
            d_optimizer.zero_grad()
            noise = torch.randn(batch_size, noise_dim).to(device)

            # Generate fake sequences
            fake_sequences = generator(noise)

            real_outputs = discriminator(real_sequences)
            fake_outputs = discriminator(fake_sequences)

            # Label smoothing for real and fake labels
            real_labels = torch.ones_like(real_outputs) * 0.9  # Label smoothing for real
            fake_labels = torch.zeros_like(fake_outputs)  # No smoothing for fake

            d_loss_real = criterion(real_outputs, real_labels)
            d_loss_fake = criterion(fake_outputs, fake_labels)
            d_loss = (d_loss_real + d_loss_fake) / 2

            d_loss.backward(retain_graph=True)  # Retain the graph for the next backward pass
            d_optimizer.step()

            # Train Generator
            g_optimizer.zero_grad()
            fake_outputs = discriminator(fake_sequences)
            g_loss = criterion(fake_outputs, torch.ones_like(fake_outputs))

            g_loss.backward()
            g_optimizer.step()

            total_d_loss += d_loss.item()
            total_g_loss += g_loss.item()
            num_batches += 1

        # Calculate and store average losses
        avg_d_loss = total_d_loss / num_batches
        avg_g_loss = total_g_loss / num_batches
        d_losses.append(avg_d_loss)
        g_losses.append(avg_g_loss)

        # Print progress
        print(f"Epoch [{epoch+1}/{num_epochs}] - D Loss: {avg_d_loss:.4f}, G Loss: {avg_g_loss:.4f}")

    return d_losses, g_losses


# Train the GAN
num_epochs = 500
d_losses, g_losses = train_gan(generator, discriminator, dataloader, num_epochs,
                              g_optimizer, d_optimizer, criterion, device, noise_dim)


# Plot final losses
plt.figure(figsize=(12, 6))
plt.plot(d_losses, label='Discriminator Loss', alpha=0.7)
plt.plot(g_losses, label='Generator Loss', alpha=0.7)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('GAN Training Loss Over Time')
plt.legend()
plt.grid(True)
plt.show()

# Save losses to CSV
loss_df = pd.DataFrame({
    'Epoch': range(1, num_epochs + 1),
    'Discriminator_Loss': d_losses,
    'Generator_Loss': g_losses
})
loss_df.to_csv('gan_training_losses.csv', index=False)
print("\nLosses have been saved to 'gan_training_losses.csv'")

# Generate some sample sequences
with torch.no_grad():
    test_noise = torch.randn(5, noise_dim).to(device)
    generated_sequences = generator(test_noise)

# Inverse transform and plot the generated sequences
generated_sequences = generated_sequences.cpu().numpy()
generated_sequences = scaler.inverse_transform(generated_sequences.reshape(-1, 1))

plt.figure(figsize=(10, 6))
plt.plot(generated_sequences[:50], label='Generated Sequence', color='r')
plt.title("Generated Sequence")
plt.show()
```
